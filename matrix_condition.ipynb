```python
# Generate random 3 tall vector

def generate_single_vector(height):
    ''' Param: How tall the vector should be '''
    
    # Random numbers in this vector are in range [-10, 10]
    nums = []
    for i in range(-10, 11):
        nums.append(i)
 
    # Create the vector based on these random numbers
    v = []
    for i in range(0, height):
        v.append(random.choice(nums))
        
    return vector(QQ, v)
        
```

```python
# Generate a dozen 'height' tall vectors

def generate_dozen(height):
    vectors = []

    for i in range(0, 12):
        vectors.append(generate_single_vector(height))

    return vectors
```

```python
# Estimate condition numbers based on vectors:
def estimate_condition(A, vectors):
    print("-----------------------")
    print("Conditions for matrix:") 
    print(A)
    print("-----------------------")
    
    for i in range(0, len(vectors)):
        w = A*vectors[i]
        print(w.norm().n())
```

```python
# Put all the pieces together for 3x3 matrix
def solve_3x3_matrix():
    print("#####################################################")
    A = generate_random_3x3()
    vectors = generate_dozen(height=3)
    print("Initial vectors:")
    print(vectors)
    estimate_condition(A, vectors)
```

```python
# Repeat 5x

for i in range(0, 5):
    solve_3x3_matrix()
```
